```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y = np.sin(x)

plt.plot(x,y)
plt.figure(figsize=(6,4))
plt.show()
```

```python



y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin", color='b')
plt.plot(x,y2,label="cos", color='r')
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend(loc=0)
plt.show
```
