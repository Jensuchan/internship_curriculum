```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,10,0.01)
y = 3*x + 5
plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.show()

y_noise = y + np.random.randn(len(y))
plt.figure(figsize=(6,4))
plt.plot(x, y_noise)
plt.show()

coefficients = np.polyfit(x,y_noise,1)
print("최적상수>>>",coefficients)

f1 = np.poly1d(coefficients)
plt.plot(x, f1(x))
plt.show()
```
