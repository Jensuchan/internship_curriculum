```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("./train.csv", encoding="utf-8")

plt.rc('font', family='Malgun Gothic')
x = df['AGE']
y = df['NOX']

coefficients = np.polyfit(x, y, 1) # x y  1()  => Numpy  y=mx+n   [m,n] 
poly1d = np.poly1d(coefficients) #     1()

plt.xlabel('1940    ')
plt.ylabel(' ')
plt.title('1940          ')

plt.plot(x,y, 'ob', label ="scatterplot")
plt.plot(x, poly1d(x), 'r-', label='trendline')
plt.legend(loc=0)
```
