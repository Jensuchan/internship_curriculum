```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("./train.csv", encoding="utf-8")

plt.rc('font', family='Malgun Gothic')
x = df['AGE']
y = df['NOX']

coefficients = np.polyfit(x, y, 1) # x와 y에 대해 1차다항식(선형)으로 구성 => Numpy 배열로 y=mx+n 에서 최적의 [m,n]값을 받는다
poly1d = np.poly1d(coefficients) # 최적의 상수를 받아서 작성한 1차함수(추세선)

plt.xlabel('1940년 이전에 건축된 자가주택의 비율')
plt.ylabel('일산화질소의 농도')
plt.title('1940년 이전에 건축된 자가주택의 비율에 대한 일산화질소 농도에 관한 산점도 그래프')

plt.plot(x,y, 'ob', label ="scatterplot")
plt.plot(x, poly1d(x), 'r-', label='trendline')
plt.legend(loc=0)
```
